```python
#Import the digit data:
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digits = load_digits(n_class = 3)
digits.data.shape
```

```python
#?load_digits
```

```python
#Visualizing the Data
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw={'xticks':[], 'yticks':[]},
                             gridspec_kw=dict(hspace=0.1, wspace=0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap='binary', interpolation='nearest',
                  clim=(0, 16))
plot_digits(digits.data)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(digits.data, digits.target, test_size=0.20, random_state=0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2).fit(digits.data)
components = pca.transform(digits.data)
projected = pca.inverse_transform(components)
```

```python
plt.scatter(components[:,0],components[:,1], c=digits.target)
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda =LDA(n_components = 2)
lda.fit(xtrain,ytrain)
yn=lda.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest, yn)
```

```python
ldatransform = lda.transform(xtrain)
```

```python
plt.scatter(ldatransform[:,0],ldatransform[:,1], c=ytrain)
plt.show()
```

```python
#?plt.scatter
```

```python
pca = PCA(n_components = 2).fit(xtrain)
components = pca.transform(xtrain)
```

```python
plt.scatter(components[:,0],components[:,1], c=ytrain)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
np.set_printoptions(precision=4)
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
wine = load_wine()
X = np.array(wine.data)
y = np.array(wine.target)
X
```

```python
y.shape
```

```python
X[160]
```

```python
#Applying PCA on Wine Dataset:
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # creates an instance of PCA class
results = pca.fit(X) # applies PCA on predictor variables
Z = results.transform(X) # create a new array of latent variables
plt.scatter(Z[:,0],Z[:,1], c=y)
plt.show()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda =LDA(n_components = 2)
X_lda = lda.fit_transform(X, y)
plt.scatter(X_lda[:,0],X_lda[:,1], c=y)
plt.show()
```
