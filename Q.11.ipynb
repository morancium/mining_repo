```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
rno = 20310007
```

```python
ld = rno%10
ld
```

```python
sld = rno%100//10
sld
```

```python
X = np.random.uniform(5+ld,20+ld,100000000)
```

```python
Y = np.random.uniform(50+sld,80+sld,100000000)
```

```python
U = Y + X
```

```python
V = Y - X
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(U, bins = 100, density = True)
 
# Show plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(V, bins = 100)
 
# Show plot
plt.show()
```
