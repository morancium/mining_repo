## Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.svm import SVC
```

```python
rawdata = pd.read_csv("advertising.csv")
rawdata.shape
```

```python
rawdata.head(10)
```

```python
del_columns = ["Ad Topic Line","City","Country","Timestamp"]
rawdata.drop(columns = del_columns, inplace = True)
x = rawdata.drop(columns = "Clicked on Ad")
x.head(10)
```

```python
# x=(x-x.mean())/x.std()
```

```python
# x.head(10)
```

```python
y = rawdata["Clicked on Ad"].copy()
y.head(10)
```

```python
c=[0.01,0.1,1,10,100,1000] # different values of c parameter
```

## SVM Models
### 1) Linear Kernel 

```python
for i in c:
    model = SVC(kernel = 'linear' ,C = i)
    model.fit(x,y)
    pred = model.predict(x)
    SupportVector = model.support_vectors_
    accuracy = model.score(x, y)
    print(f"Value of C: {i}")
    print(f"Confusion matrix: \n{confusion_matrix(pred,y)}")
    print(f"Number of Support Vector: {SupportVector.size}")
    print(f"Accuracy of the model: {accuracy}\n\n")
```

### 2) Polynomial Kernel 

```python
for i in c:
    model = SVC(kernel = 'poly',degree = 2,C = i)
    model.fit(x,y)
    pred = model.predict(x)
    SupportVector = model.support_vectors_
    accuracy = model.score(x, y)
    print(f"Value of C: {i}")
    print(f"Confusion matrix: \n{confusion_matrix(pred,y)}")
    print(f"Number of Support Vector: {SupportVector.size}")
    print(f"Accuracy of the model: {accuracy}\n\n")
```

### 3) RBF Kernel

```python
for i in c:
    model = SVC(kernel = 'rbf',C = i)
    model.fit(x,y)
    pred = model.predict(x)
    SupportVector = model.support_vectors_
    accuracy = model.score(x, y)
    print(f"Value of C: {i}")
    print(f"Confusion matrix: \n{confusion_matrix(pred,y)}")
    print(f"Number of Support Vector: {SupportVector.size}")
    print(f"Accuracy of the model: {accuracy}\n\n")
```
