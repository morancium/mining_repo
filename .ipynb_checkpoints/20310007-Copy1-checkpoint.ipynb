#### Importing Libraries and Dataset

```python
import time
import scipy
import numpy as np
import cv2 as cv
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

#### Importing Image

```python
myImg = io.imread("20111015.jpg")
```

```python
#Original shape of the image
myImg.shape
```

```python
#Original Image
plt.imshow(myImg)
plt.show()
```

```python
#Flattening the image into a (Len*breth,3) matrix
myImgnew = myImg.reshape(myImg.shape[0]*myImg.shape[1],myImg.shape[2])
myImgnew.shape
```

### KNN Clustering

```python
def KNN_image_cluster(path, number_of_cluster=5):
    myImg = io.imread(path)
    plt.imshow(myImg)
    myImgnew = myImg.reshape(myImg.shape[0]*myImg.shape[1],myImg.shape[2])
    kmeans = KMeans(n_clusters=number_of_cluster, max_iter=10000,random_state = 170423)
    begin = time.time()
    kmeans.fit(myImgnew)
    end = time.time()
    labels = kmeans.predict(myImgnew)
    Ccentres = kmeans.cluster_centers_.astype(int)
    print("Center of clusters: \n",Ccentres)
    newimage = []
    #Replacing each pixel to nearest Cluster center
    for i in range(len(labels)):
        newimage.append(Ccentres[labels[i]])
    newimage = np.array(newimage)
    #Reshaping the image back to original shape
    newimagesee = newimage.reshape(myImg.shape[0],myImg.shape[1],myImg.shape[2])
    newimage = np.array(newimagesee)
    print(f"Time Taken for Clustering: {np.round(end - begin,4)}s")
    plt.imshow(newimage)
    plt.show()
```

```python
# image using only 5 colors
KNN_image_cluster("20111015.jpg",5)
```

```python
# image using only 10 colors
KNN_image_cluster("20111015.jpg",10)
```

```python
# image using only 20 colors
KNN_image_cluster("20111015.jpg",20)
```

### GMM Clustering

```python
def GMM_image_cluster(path, number_of_cluster=5):
    myImg = io.imread(path)
    myImgnew = myImg.reshape(myImg.shape[0]*myImg.shape[1],myImg.shape[2])
    gmm = GaussianMixture(n_components=number_of_cluster, covariance_type='full', random_state=0)
    begin = time.time()
    out=gmm.fit(myImgnew)
    end = time.time()
    labels = gmm.predict(myImgnew)
    centers = np.empty(shape=(out.n_components, myImgnew.shape[1]))
    #finding the cemter of each cluster
    for i in range(out.n_components):
        density = scipy.stats.multivariate_normal(cov=out.covariances_[i], mean=out.means_[i]).logpdf(myImgnew)
        centers[i, :] = myImgnew[np.argmax(density)]
    centers=centers.astype(int)
    print("Center of clusters: \n",centers)
    newimage = []
    #Replacing each pixel to nearest Cluster center
    for i in range(len(labels)):
        newimage.append(centers[labels[i]])
    newimage = np.array(newimage)
    #Reshaping the image back to original shape
    newimagesee = newimage.reshape(myImg.shape[0],myImg.shape[1],myImg.shape[2])
    newimage = np.array(newimagesee)
    print(f"Time Taken for Clustering: {np.round(end - begin,4)}s")
    plt.imshow(newimage)
    plt.show()
```

```python
# image using only 5 colors
GMM_image_cluster("20111015.jpg",5)
```

```python
# image using only 10 colors
GMM_image_cluster("20111015.jpg",10)
```

```python
# image using only 20 colors
GMM_image_cluster("20111015.jpg",20)
```
